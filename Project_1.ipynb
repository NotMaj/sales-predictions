```python
#GRAPHS
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df.columns
```

```python
#Histogram to view the distributions of various features in your dataset.
# Item type sold
item_count = df['Item_Type'].value_counts().sort_values(ascending=False)
item_count




```

```python
item_count.plot(kind='barh', grid=False)
plt.xlabel('Number Sold')
plt.ylabel('Food Type')

plt.style.use('seaborn');
```

```python
#Item_Outlet_Sales
ax= sns.histplot(df['Item_Outlet_Sales'], bins ='auto', color='green')
ax.set(xlabel ='Sales',
        title='Outlet Sales')

ax.axvline(df['Item_Outlet_Sales'].median(), color='k', ls=':'
,label=f"Median Price={round(df['Item_Outlet_Sales'].median(),2)}")

plt.legend()
plt.show();
```

```python
bx= sns.histplot(df['Item_Weight'], bins='auto', color='orange')
bx.set(xlabel='Item Weights',
        title ='Distribution of Item Weights')
bx.axvline(df['Item_Weight'].median(),
            ls=':',
            color='black',
            label =f"Median Weight ={round(df['Item_Weight'].median(),2)}")
plt.legend()
plt.show();
```
