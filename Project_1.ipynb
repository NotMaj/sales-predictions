```python
#Histogram to view the distributions of various features in your dataset.
# Item type sold
sns.countplot(data=df, y='Item_Type')
plt.ylabel('Item Types')
plt.xlabel('Items Sold')
plt.title('Number of items sold per type');

```

```python
#Item_Outlet_Sales
ax= sns.histplot(df['Item_Outlet_Sales'], bins ='auto', color='green')
ax.set(xlabel ='Sales',
       ylabel='Items Sold',
        title='Distribution of Outlet Sales')

ax.axvline(df['Item_Outlet_Sales'].median(), color='k', ls=':'
,label=f"Median Price={round(df['Item_Outlet_Sales'].median(),2)}")

plt.legend()
plt.show();
```

```python
year_filter = df['Outlet_Establishment_Year'] == 2009
outlet_filter =df['Outlet_Location_Type'] == 'Tier 3'
temp_df = df.loc[year_filter & outlet_filter]

```

```python
sns.lmplot(data=temp_df, x='Item_Outlet_Sales', y='Item_Weight', hue='Item_Fat_Content')
plt.ylabel('Item Weights')
plt.xlabel('Outlet Sales');
```

The scatterplot which highlights the multivariate analysis of 2 variables of item sales and weight. The plot shows that the variables have no correlation or relationship with each other.

```python
year = df.groupby('Outlet_Establishment_Year').mean()

#Create figure with 2 subplots
fig, axes = plt.subplots(2,1, figsize=(15,10))

axes[0].plot(year.index, year['Item_Weight'])
axes[0].set_title('Item Weight', fontsize=20)
axes[0].set_ylabel('Average Item Weights', fontsize=20)
axes[0].set_xlabel("Year", fontsize=20)
axes[0].set_xticks(year.index)
axes[0].grid()

axes[1].plot(year.index, year['Item_MRP'])
axes[1].set_title('Average Item MRP', fontsize=20)
axes[1].set_ylabel('Item Mrp', fontsize=20)
axes[1].set_xlabel("Year", fontsize=20)
axes[1].set_xticks(year.index)
axes[1].grid()

plt.tight_layout();
```
